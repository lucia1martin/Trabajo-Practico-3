```python
import pandas as pd
import numpy as np

# ================================
# 1) Partimos de tu base
# ================================
columnas_nuevas = [
    "codusu", "nro_hogar", "ano4",
    "sexo", "años_cumplidos", "estado_civil", "cobertura_medica",
    "nivel_ed", "estado", "inactividad", "ipcf"
]
df_hogar = df_interes[columnas_nuevas].copy()

# ================================
# 2) Tabla de adulto equivalente
# ================================
# El archivo trae encabezado sucio y una fila 'Edad' que hay que sacar
tabla = (pd.read_excel("tabla_adulto_equiv.xlsx", skiprows=2)
           .rename(columns={"Unnamed: 0":"edad_label","Unnamed: 1":"Mujer","Unnamed: 2":"Varón"}))

# eliminar filas sin etiqueta y la fila de encabezado 'Edad'
tabla = tabla.dropna(subset=["edad_label"]).copy()
tabla["edad_label"] = tabla["edad_label"].astype(str).str.strip()
tabla = tabla[tabla["edad_label"] != "Edad"]

# formato largo
tabla_long = tabla.melt(
    id_vars="edad_label",
    value_vars=["Mujer", "Varón"],
    var_name="sexo_tabla",
    value_name="adulto_equiv"
)
tabla_long["adulto_equiv"] = pd.to_numeric(tabla_long["adulto_equiv"], errors="coerce")

# ================================
# 3) Normalizadores
# ================================
def normalizar_sexo_valor(x):
    # acepta 1/2, 1.0/2.0, varon/varón/hombre/m, mujer/f, etc.
    try:
        xn = float(x)
        if int(xn) == 1: return "Varón"
        if int(xn) == 2: return "Mujer"
    except Exception:
        pass
    s = str(x).strip().lower()
    if s in {"1","1.0","varon","varón","hombre","masculino","m","v"}: return "Varón"
    if s in {"2","2.0","mujer","femenino","f"}:                       return "Mujer"
    return pd.NA

def edad_a_tramo(e):
    # OJO: las etiquetas deben coincidir EXACTO con la tabla (p.ej., "1año" SIN espacio)
    if pd.isna(e): return pd.NA
    try:
        e = int(np.floor(float(e)))
    except Exception:
        return pd.NA
    if e < 1:            return "Menor de 1 año"
    if e == 1:           return "1año"           # <-- cambio clave: SIN espacio
    if 2  <= e <= 14:    return f"{e} años"
    if e == 15:          return "15 años"
    if e == 16:          return "16 años"
    if e == 17:          return "17 años"
    if 18 <= e <= 29:    return "18 a 29 años"
    if 30 <= e <= 45:    return "30 a 45 años"
    if 46 <= e <= 60:    return "46 a 60 años"
    if 61 <= e <= 75:    return "61 a 75 años"
    return "más de 75 años"  # con tilde, igual que en el Excel

# ================================
# 4) Sexo y edad en tu base
# ================================
df_hogar["sexo_std"]   = df_hogar["sexo"].apply(normalizar_sexo_valor)
df_hogar["edad_label"] = pd.to_numeric(df_hogar["años_cumplidos"], errors="coerce").apply(edad_a_tramo)

# ================================
# -------------------------
# 5) Merge exacto por etiquetas (edad_label, sexo)
# -------------------------
df_hogar = df_hogar.merge(
    tabla_long.rename(columns={"sexo_tabla":"sexo_ref"}),
    left_on=["edad_label", "sexo_std"],
    right_on=["edad_label", "sexo_ref"],
    how="left",
    suffixes=("", "_tabla")   # conserva tus columnas originales
).drop(columns=["sexo_ref"], errors="ignore")

# ================================
# 6) Suma por hogar (preserva NaN)
# ================================
df_hogar["ad_equiv_hogar3"] = (
    df_hogar.groupby(["codusu","nro_hogar"])["adulto_equiv"]
            .transform(lambda s: s.sum(min_count=1))
)
# ================================
# 7) Diagnóstico rápido
# ================================
print("Con coeficiente asignado:", df_hogar["adulto_equiv"].notna().sum(), "/", len(df_hogar))
print("Valores únicos de edad_label que no machearon (si los hubiera):")
print(df_hogar.loc[df_hogar["adulto_equiv"].isna(),"edad_label"].dropna().value_counts().head(10))
print("Valores únicos de sexo_std que no machearon (si los hubiera):")
print(df_hogar.loc[df_hogar["adulto_equiv"].isna(),"sexo_std"].dropna().value_counts().head(10))

# chequeo visual
# Vista de control con el nombre correcto
print(df_hogar[[
    "codusu","nro_hogar","sexo","sexo_std",
    "años_cumplidos","edad_label","adulto_equiv","ad_equiv_hogar3"
]].head(20))

```

```python
#nueva intentando que no tenga tantos NaNs.
import re
import numpy as np
import pandas as pd
import statsmodels.api as sm
from IPython.display import display

def drop_one_dummy_per_group(X, min_levels=('NIVEL_ED','CH07')):

    #esto lo hacemos para que la dummie no traiga Nan.
    X = X.copy()
    # agrupar columnas que parecen dummies: prefijo_sufijo
    pat = re.compile(r'^(.+?)_([^_]+)$')
    groups = {}
    for c in X.columns:
        m = pat.match(c)
        if m:
            pref, suf = m.group(1), m.group(2)
            groups.setdefault(pref, []).append((c, suf))

    to_drop = []
    for pref, items in groups.items():
        # solo si hay más de una dummy del mismo prefijo
        if len(items) > 1:
            # si pediste referencias explícitas
            if pref in min_levels:
                # elige la columna cuyo nombre termina en "_1" si existe, sino la de menor orden
                cand = [c for c,s in items if s == '1']
                if cand:
                    to_drop.append(cand[0])
                    continue
            # por defecto: dropear la de menor sufijo alfanumérico
            items_sorted = sorted(items, key=lambda x: x[1])
            to_drop.append(items_sorted[0][0])

    return X.drop(columns=[c for c in to_drop if c in X.columns], errors='ignore'), to_drop

#  1) Mantener numéricas/booleanas
X = X_train.select_dtypes(include=["number", "boolean"]).copy()

# limpiar columnas constantes/duplicadas
const_cols = [c for c in X.columns if X[c].nunique(dropna=True) <= 1]
if const_cols:
    X = X.drop(columns=const_cols)
X = X.T.drop_duplicates().T

# reemplazos y tipos
X = X.replace([np.inf, -np.inf], np.nan).astype("float64")
y = pd.to_numeric(y_train, errors="coerce").astype("float64")

#  2) sacar una dummy por grupo (evitar dummy trap)
X, dropped = drop_one_dummy_per_group(X, min_levels=('NIVEL_ED','CH07'))

#  3) filtrar filas válidas
valid = (~X.isna().any(axis=1)) & (~y.isna())
X, y = X.loc[valid], y.loc[valid]

#  4) agregar constante
if "const" not in X.columns:
    X = sm.add_constant(X, has_constant="add").astype("float64")



#  5) estimar Logit
modelo = sm.Logit(y, X).fit(disp=False, maxiter=200)

tabla = pd.DataFrame({
    "Coeficiente": modelo.params,
    "Error estándar": modelo.bse,
    "Odds Ratio": np.exp(modelo.params)
}).round(4)

print("=== Regresión Logística usando X_train (sin dummy trap) ===")
print("Referencia (dummies dropeadas):", dropped)
display(tabla.head(20))
```

```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, balanced_accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# seleccionamos columnas numericas
num_cols = [c for c in X_train.columns if c != 'const']

Xtr = X_train[num_cols].copy()
Xte = X_test[num_cols].copy()
ytr = y_train.copy()
yte = y_test.copy()

#  Convertimos a float64 
Xtr = Xtr.apply(pd.to_numeric, errors='coerce').astype(float)
Xte = Xte.apply(pd.to_numeric, errors='coerce').astype(float)

# KNN
def knn_pipe(k):
    return Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler()),
        ('knn', KNeighborsClassifier(n_neighbors=k, weights='uniform', metric='minkowski', p=2))
    ])

res = []
cms = {}

for k in [1, 5, 10]:
    pipe = knn_pipe(k)
    pipe.fit(Xtr, ytr)
    y_pred = pipe.predict(Xte)

    res.append({
        'K': k,
        'accuracy': accuracy_score(yte, y_pred),
        'balanced_acc': balanced_accuracy_score(yte, y_pred),
        'precision_1': precision_score(yte, y_pred, pos_label=1, zero_division=0),
        'recall_1': recall_score(yte, y_pred, pos_label=1, zero_division=0),
        'f1_1': f1_score(yte, y_pred, pos_label=1, zero_division=0),
    })
    cms[k] = confusion_matrix(yte, y_pred)

#  Resultados 
tabla_knn = pd.DataFrame(res).sort_values('K').reset_index(drop=True)
print(tabla_knn.round(3))

for k in [1, 5, 10]:
    print(f"\nMatriz de confusión (K={k}):\n", cms[k])
```

```python
# === KNN con K-CV (5 folds) sobre respondieron_2025 ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# -------------------- 1) Preparación --------------------
df = respondieron_2025.copy()

if "pobre" not in df.columns:
    raise KeyError("No encuentro la columna 'pobre' en respondieron_2025. Agregala antes de correr esto.")

# Solo numéricas/booleanas para X
X = df.select_dtypes(include=["number", "boolean"]).drop(columns=["pobre"], errors="ignore").copy()
y = df["pobre"].astype("int64")

# (Opcional) Info rápida para verificar que hay datos
print(f"[INFO] Filas totales: {len(df)} | Columnas X: {X.shape[1]}")
print(f"[INFO] Balance de y (pobre):\n{y.value_counts(dropna=False)}\n")

# -------------------- 2) Cross-Validation --------------------
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)

k_values = list(range(1, 32, 2))  # K impares 1..31
acc_mean, acc_std = [], []

for k in k_values:
    pipe = Pipeline([
        ("imp",   SimpleImputer(strategy="median")),
        ("sc",    StandardScaler()),
        ("knn",   KNeighborsClassifier(n_neighbors=k))
    ])
    scores = cross_val_score(pipe, X, y, cv=cv, scoring="accuracy", n_jobs=None)
    acc_mean.append(scores.mean())
    acc_std.append(scores.std())

# Mejor K
best_idx = int(np.argmax(acc_mean))
best_k = k_values[best_idx]
best_acc = acc_mean[best_idx]
print(f"Mejor K por accuracy CV: {best_k} | accuracy CV media = {best_acc:.3f}")

# -------------------- 3) Entrenar modelo final con best_k --------------------
modelo_knn_kcv = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("sc",  StandardScaler()),
    ("knn", KNeighborsClassifier(n_neighbors=best_k))
])
modelo_knn_kcv.fit(X, y)
print("[OK] Modelo 'KNN con K-CV' entrenado con todo el set usando K* =", best_k)

# -------------------- 4) Gráfico --------------------
plt.figure(figsize=(7,5))
plt.plot(k_values, acc_mean, marker="o")
plt.fill_between(k_values,
                 np.array(acc_mean) - np.array(acc_std),
                 np.array(acc_mean) + np.array(acc_std),
                 alpha=0.3)
plt.title("Curva de validación para K en KNN")
plt.xlabel("K vecinos")
plt.ylabel("Accuracy CV (media 5-fold)")
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import (
    confusion_matrix, ConfusionMatrixDisplay,
    accuracy_score, roc_auc_score, roc_curve
)

# --- Funciones auxiliares ---
def only_numeric(df):
    return df.select_dtypes(include=["number", "boolean"]).copy()

def metrics_accuracy_auc(y_true, y_score, y_pred, name):
    acc = accuracy_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_score)
    print(f"{name:>14} | Accuracy: {acc:.3f} | AUC: {auc:.3f}")
    return acc, auc

# --- Paso inicial ---
Xtr = only_numeric(X_train).copy()
Xte = only_numeric(X_test).copy()
ytr = pd.to_numeric(y_train, errors="coerce").astype(int)
yte = pd.to_numeric(y_test,  errors="coerce").astype(int)

# Alinear columnas por si difieren en orden o falta alguna
Xte = Xte.reindex(columns=Xtr.columns, fill_value=np.nan)

# --- 1) KNN: elegir K óptimo en TRAIN (CV) ---
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
k_values = list(range(1, 32, 2))

pipe_knn = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("sc",  StandardScaler()),
    ("knn", KNeighborsClassifier())
])

acc_cv = []
for k in k_values:
    pipe_knn.set_params(knn__n_neighbors=k)
    scores = cross_val_score(pipe_knn, Xtr, ytr, cv=cv, scoring="accuracy")
    acc_cv.append(scores.mean())


# Forzar K óptimo = 19 para coherencia con todo el TP
best_k = 19
pipe_knn.set_params(knn__n_neighbors=best_k).fit(Xtr, ytr)

# --- 2) LOGIT (sklearn) en TRAIN ---
pipe_logit = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("sc",  StandardScaler()),
    ("lr",  LogisticRegression(solver="liblinear", max_iter=1000))
]).fit(Xtr, ytr)

# --- 3) Predicciones en TEST (umbral 0.5) ---
proba_knn   = pipe_knn.predict_proba(Xte)[:, 1]
proba_logit = pipe_logit.predict_proba(Xte)[:, 1]

pred_knn    = (proba_knn  >= 0.5).astype(int)
pred_logit  = (proba_logit >= 0.5).astype(int)

# --- 4) Métricas (Accuracy y AUC) ---
print(f"[KNN] Mejor K fijado: {best_k} (óptimo CV anterior)\n")
print("Modelo          | Accuracy | AUC")
print("----------------+----------+-----")
acc_knn,  auc_knn  = metrics_accuracy_auc(yte, proba_knn,  pred_knn,  f"KNN (K={best_k})")
acc_log,  auc_log  = metrics_accuracy_auc(yte, proba_logit, pred_logit, "Logit")

# --- 5) Matrices de confusión (umbral 0.5) ---
fig, axes = plt.subplots(1, 2, figsize=(9,4), constrained_layout=True)
for ax, y_pred, title in zip(
    axes,
    [pred_knn, pred_logit],
    [f"KNN (K={best_k})", "Logit"]
):
    cm = confusion_matrix(yte, y_pred)
    disp = ConfusionMatrixDisplay(cm, display_labels=["No pobre (0)","Pobre (1)"])
    disp.plot(ax=ax, values_format="d", cmap="Blues", colorbar=False)
    ax.set_title(f"Matriz de confusión — {title}")
plt.show()

# --- 6) Curva ROC (ambos modelos) ---
fpr_knn, tpr_knn, _ = roc_curve(yte, proba_knn)
fpr_log, tpr_log, _ = roc_curve(yte, proba_logit)

plt.figure(figsize=(6,5))
plt.plot(fpr_knn, tpr_knn, label=f"KNN (AUC={auc_knn:.3f})")
plt.plot(fpr_log, tpr_log, label=f"Logit (AUC={auc_log:.3f})")
plt.plot([0,1],[0,1], linestyle="--", linewidth=1)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curvas ROC — Test")
plt.legend()
plt.tight_layout()
plt.show()
```
